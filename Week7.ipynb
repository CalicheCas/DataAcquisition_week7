# Grouping and Combining Datas with Pandas

### Load the ratings by user information that you collected into a pandas dataframe.

```python
import pandas as pd
df = pd.read_csv("MovieRatings.csv")
df2 = df
df3 = df
df
```

### Show the average ratings for each user and each movie
#### 1. Average Rating of each user

```python
df["Ave Rating"] = df.sum(axis=1) / (len(df.columns) - df.isnull().sum(axis=1) - 1)
df.round(2)
```

#### Average Rating of Each Movie


```python
df = df.append(df.agg({'Joker' : ['sum', 'mean'], 'Frozen' : ['sum', 'mean'], 'IT' : ['sum', 'mean'], 'Breaking_Bad' : ['sum', 'mean'], 'Jason_Bourne' : ['sum', 'mean'], 'Mission_Impossible' : ['sum', 'mean'], 'Ave Rating' : ['sum', 'mean'], }))
df.round(2)
```

### Create a new pandas dataframe, with normalized ratings for each user. Again, show the average
### ratings for each user and each movie.

```python
def normalize(base, min, max):
    divider = base - min
    denom = max - min

    return divider / denom
```

```python
ocols = df2.columns
# Calculating min and max of each row to use in data normalization equation
df2['min'] = df2.min(axis=1)
df2['max'] = df2.max(axis=1)
```

```python
df2.round(2)
```

### Creating new normalized dataframe

```python
new_df = pd.DataFrame(columns=ocols)
print(new_df)
group = df2.groupby("Rater_Name")

# Itearete over every person's ratings
for person, movie_df in group:
    
    name = person
    min = movie_df["min"].values[0]
    max = movie_df["max"].values[0]

    # Itearate over movie titles
    for col in movie_df.columns:
        # Ignore rater name min and max columns
        if col not in ["Rater_Name", "min", "max"]:        
            # Get current rating
            base = movie_df[col].values[0]
        
            # Calculate normalize rating
            if base > 0:
                nr = normalize(base, min, max)
                #Update Cell value
                new_df.loc[person, col] = nr
                
new_df
```

### Provide a text-based conclusion: explain what might be advantages and disadvantages of using normalized ratings instead of the actual ratings.

The advange of normalizing data is providing a simpler data set for easier analysis. Since the range of values are between 0 and 1 it also provides better context of the overal data set. It will also allow for more meaningful comparissons between data sets that do not have as many data points.

On the other hand normalizing data does not work well with small data sets. Also, if all values in the data set are the same, minumum and maximum values will be offset the calculation, which will end up in a NaN entry.

#### Data Standardization

```python
def standized(base, mean, variance):
    
    return (base - mean) / variance     
```

```python
df3['mean'] = df3.mean(axis=1)
df3['var'] = df3.var(axis=1)
df3   
```

```python
std_df = pd.DataFrame(columns=ocols)
group = df3.groupby("Rater_Name")

# Itearete over every person's ratings
for person, movie_df in group:
    
    name = person
    mean = movie_df["mean"].values[0]
    var = movie_df["var"].values[0]

    # Itearate over movie titles
    for col in movie_df.columns:
        # Ignore rater name min and max columns
        if col not in ["Rater_Name", "mean", "var", "min", "max"]:        
            # Get current rating
            base = movie_df[col].values[0]
        
            # Calculate normalize rating
            if base > 0:
                std = standized(base, mean, var)
                #Update Cell value
                std_df.loc[person, col] = std
```
